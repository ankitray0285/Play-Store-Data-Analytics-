# Time-Based Graph Visibility (3 PM - 5 PM Restriction)

```python
#import required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Load data
df = pd.read_csv("Play Store Data.csv")
```

```python
#Define a Function to Check the Time Window
def is_within_time_range():
    """
    Function to check if the current time is between 3 PM and 5 PM IST.
    """
    now = datetime.datetime.now()
    current_hour = now.hour
    return 15 <= current_hour < 17  # 15:00 to 17:00 in 24-hour format
```

```python
#Apply Time Restriction to Graphs
def show_graph():
    """
    Function to display the graph only within the specified time range.
    """
    if is_within_time_range():
        # Sample visualization: Top 5 app categories by count
        top_categories = df['Category'].value_counts().head(5)
        plt.figure(figsize=(10, 5))
        top_categories.plot(kind='bar', color='skyblue')
        plt.title("Top 5 App Categories")
        plt.xlabel("Category")
        plt.ylabel("Number of Apps")
        plt.xticks(rotation=45)
        plt.show()
    else:
        print("❌ Graphs are only visible between 3 PM and 5 PM IST.")
```

```python
#Run the Function
show_graph()
```

# How It Works
The function is_within_time_range() checks if the current time falls between 15:00 (3 PM) and 17:00 (5 PM) IST.

If within the time, the visualization is displayed.

Otherwise, it prints a message: "Graphs are only visible between 3 PM and 5 PM IST."
